```python
import torch
from transformers import AutoProcessor, MusicgenForConditionalGeneration
import scipy.io.wavfile as wavfile
from IPython.display import Audio  # For Jupyter notebooks
import numpy as np
```

```python
# Load model and processor
processor = AutoProcessor.from_pretrained("/home/shaikat/.cache/kagglehub/models/metaresearch/musicgen/pyTorch/small/1")
model = MusicgenForConditionalGeneration.from_pretrained("/home/shaikat/.cache/kagglehub/models/metaresearch/musicgen/pyTorch/small/1")
```

```python
# Move to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)
```

```python
# Text prompt
prompt = "8bit video game music"
```

```python
# Process the text input
inputs = processor(
    text=[prompt],
    padding=True,
    return_tensors="pt",
).to(device)
```

```python
# Generate audio
audio = model.generate(**inputs, max_new_tokens=512)
```

```python
# Convert to numpy
audio_data = audio[0].cpu().numpy()
sample_rate = model.config.audio_encoder.sampling_rate
```

```python
Audio(audio_data, rate=sample_rate)
```

```python
del model
del processor
torch.cuda.empty_cache()
torch.cuda.synchronize()
```
